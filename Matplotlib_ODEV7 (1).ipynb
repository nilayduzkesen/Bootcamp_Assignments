```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import warnings

warnings.filterwarnings('ignore')

```

```python

sales = pd.read_csv('./vgsales.csv')
sales
```

```python
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
sales.groupby('Platform')['NA_Sales'].nunique().plot(kind='bar')
plt.ylabel('Sales')
plt.title('Platform Sales')

plt.subplot(1, 2, 2)
sales.groupby('Genre')['NA_Sales'].nunique().plot(kind='bar')
plt.ylabel('Sales')
plt.title('Genre Sales')

plt.tight_layout()
plt.show()
```

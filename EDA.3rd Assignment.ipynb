```python
#ODEV 3
#eksik verileri o yıl içerisinde gözlemlenen değerlere dayanarak doldur.

years =data.YEAR.unique()
print (years)
```

```python
#basta kopyaladıgım datamın Na ıceren haline donuyorum

states_all2
fill_list = ["ENROLL", "TOTAL_REVENUE", "FEDERAL_REVENUE", 
             "STATE_REVENUE", "LOCAL_REVENUE", "TOTAL_EXPENDITURE", 
             "INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
             "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE", "GRADES_PK_G", 
             "GRADES_KG_G", "GRADES_4_G", "GRADES_8_G", "GRADES_12_G", "GRADES_1_8_G", 
             "GRADES_9_12_G", "GRADES_ALL_G"] 

for sutun_adı in fill_list:
    for year in years:
        states_all2[sutun_adı].fillna(states_all2.groupby(by='YEAR').mean()[sutun_adı].loc[year],inplace=True)
data.describe()
```

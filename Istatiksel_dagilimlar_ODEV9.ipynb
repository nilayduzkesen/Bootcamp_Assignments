```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import warnings

warnings.filterwarnings('ignore')

```

```python
#ODEV1

#dağılım = np.random.dağılım_adı([],[])
#a) Verinin histogramını çiziniz b) Q-Q grafiğini çizip yorumlayınız

#histogram

plt.figure(figsize=(5,3), dpi = 100)
dagılım = np.random.normal(36,1,500)
plt.hist(dagılım, bins = 20)
plt.show()
```

```python
#Q-Q cizelim

standart_dagilimli = np.random.normal(0, 1, 500)

dagılım.sort()
standart_dagilimli.sort()
plt.figure(figsize = (10,6),dpi =100)
plt.plot(standart_dagilimli,dagılım,"o")
plt.plot([np.min(standart_dagilimli),np.max(standart_dagilimli)],[np.min(dagılım),np.max(dagılım)],c='red')
plt.xlabel('standart_dagilim')
plt.ylabel('dagılım')

plt.show()

```
